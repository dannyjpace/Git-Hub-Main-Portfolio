## Project Overview

The aim of this project is to apply Python's Pandas library for advanced data transformation tasks. Users will enhance their skills in cleaning, combining, and extracting insights from structured datasets.

---

## Dataset Description

The datasets include:

### Main Dataset
- **professor**: Names of professors.
- **department**: Departments they belong to.
- **age**: Professors' ages.

### Supplementary Dataset
- **professor**: Names of professors (for combining datasets).
- **courses**: Courses taught by professors.

---

## Transformation Tasks

### **Task 1: Add Professor Initials**
- **Objective**: Create a column `professor_initials` containing the initials of professors' first and last names.
- **Methodology**:
  - Apply a lambda function to the `professor` column to extract initials using string operations.

---

### **Task 2: Join DataFrames**
- **Objective**: Combine the main DataFrame with the supplementary DataFrame based on the `professor` column.
- **Methodology**:
  - Use the `join` function after setting `professor` as the index for both DataFrames.

---

### **Task 3: Merge DataFrames**
- **Objective**: Merge the main DataFrame and the supplementary DataFrame on the `professor` column.
- **Methodology**:
  - Use the `merge` function to combine both DataFrames while keeping the `professor` column as the key.

---

### **Task 4: Extract Last Names**
- **Objective**: Add a column `professor_last_name` containing only the last names of professors.
- **Methodology**:
  - Use string operations to split the `professor` column and extract the last name.
